# TSVD on Localisation and Big Data Assimilation

```python
import numpy as np
import os, sys, re
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def getSVDres(backgroundType, obsNb, percent, domain):
    dirRes = "../SVDresults/{}{}_{}TSVD/{}/".format(backgroundType, obsNb, percent, domain)
    fpRes = dirRes + "results.csv"
    fpAvgRes = dirRes + "resultsAvg.csv"

    res = pd.read_csv(fpRes)
    avgRes = pd.read_csv(fpAvgRes)
    return res, avgRes

def displaySVDres(backgroundType, obsNb, percentages):
    for idx in range(len(percentages)):
        percent = percentages[idx]
        res8, res8Avg = getSVDres(backgroundType, obsNb, percent, 8)
        res6, res6Avg = getSVDres(backgroundType, obsNb, percent, 6)
        print("Subdomain {} with {}% - u_0 = {} - obsNb = {}".format(8, percent, backgroundType, obsNb))
#         display(res8)
        display(res8Avg)
        print("Subdomain {} with {}% - u_0 = {} - obsNb = {}".format(6, percent, backgroundType, obsNb))
#         display(res6)
        display(res6Avg)
```

## SVD - u_c = MeanHist - Obs Frac = 0.1 - Nb Modes = Optimal

```python
percentages = [10, 20, 40, 80]
displaySVDres("MeanHist", 0.1, percentages)
```

## SVD - u_c = Idx - Obs Frac = 0.1 - Nb Modes = Optimal

```python
percentages = [10, 20, 40, 80]
displaySVDres("Idx", 0.1, percentages)
```

## SVD - u_c = Idx - Obs Frac = 1 - Nb Modes = Optimal

```python
percentages = [10, 20, 40, 80]
displaySVDres("Idx", 1, percentages)
```
