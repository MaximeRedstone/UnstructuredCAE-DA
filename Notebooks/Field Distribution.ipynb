```python
ug8 = vtktools.vtu("../../../../../data/subdomain_8/LSBU_0_8.vtu")
locations8 = ug8.GetLocations()
scalars8 = ug8.GetScalarField('TracerGeorge')
print(len(scalars8))

ug6 = vtktools.vtu("../../../../../data/subdomain_6/LSBU_0_6.vtu")
locations6 = ug6.GetLocations()
scalars6 = ug6.GetScalarField('TracerGeorge')

def normaliseScalar(scalar):
    """ Normalise a scalar field between 0 and 1 """
    normalisedScalar = []
    maxValue = np.amax(scalar)
    minValue = np.amin(scalar)
    for value in scalar:
        normalisedValue = (value - minValue) / (maxValue - minValue)
        normalisedScalar.append(normalisedValue)

    return normalisedScalar

normScalar8 = normaliseScalar(scalars8)
print(np.max(normScalar8), np.min(normScalar8))
print(np.max(scalars8), np.min(scalars8))
```

```python
rows = 2
cols = 1
fig, ax = plt.subplots(rows, cols, sharex = False,sharey = False, figsize = (15, 15))
x = np.linspace(0, len(scalars8), len(scalars8))
axes = ax[0]
axes.plot(x, scalars8, '.b')
axes.set_xlabel('Location index', fontsize=20)
axes.set_ylabel('Scalar Value', fontsize=20)
axes.set_title('TracerGeorge Scalar Values for Subdomain 8 at Timestep 0', fontsize=20)
x = np.linspace(0, len(scalars6), len(scalars6))
axes = ax[1]
axes.plot(x, scalars6, '.b')
axes.set_xlabel('Location index', fontsize=20)
axes.set_ylabel('Scalar Value', fontsize=20)
axes.set_title('TracerGeorge Scalar Values for Subdomain 6 at Timestep 0', fontsize=20)
fig.savefig("subdomain68ScalarField0.png")
fig.show()
```
